### Dummy Variables
This technique is used to transform categorical values into a numbers to be manipulated. It consists of a matrix of 0's and a 1 in the position of the categorical value that the variable represents.

```python
# Import libraries
import pandas as pd
```

```python
# Import the dataser
absenteeism_df = pd.read_csv("Absenteeism_at_work.csv", sep = ";")

# Print the first 5 lines of the dataframe
absenteeism_df.head()
```

```python
# Transform a categorical value to dummy variables
dummy_var = pd.get_dummies(absenteeism_df["Day of the week"], prefix="Day")

# Add dummy_var to the dataframe
absenteeism_df = pd.concat([absenteeism_df, dummy_var], axis = 1)

# Remove old column from dataframe
absenteeism_df = absenteeism_df.drop("Day of the week", axis = 1)

# Print the first 5 lines to see the changes
absenteeism_df.head()
```
